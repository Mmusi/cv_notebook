```python
import matplotlib.pyplot as plt
# Pie chart
labels = ['Programming', 'Databases(SQL)', 'Excel', 'End-User Support','Data Analysis','Data Visualisation','Data Modelling']
sizes = [9, 8, 8, 10, 0.5, 2, 0.5]
#colors
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', pctdistance=0.85, startangle=90)
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Pie chart
labels = ['Programming', 'Databases(SQL)', 'Excel', 'End-User Support','Data Visualisation','Data Analysis']
sizes = [9, 8, 8, 7, 2, 0.5]
#colors
colors = ['#006D4F','#5E0DAC','#00A779','#ffcc99','#ff6666','#66b3ff']
#explsion
explode = (0.05,0.05,0.05,0.05,0.04,0.10)
 

#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, explode = explode)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```
